Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y,c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('x vs y')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2, 2, 1)  # Subplot 1
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)  # Subplot 2
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  # Subplot 3
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  # Subplot 4
plt.plot(x, y4)
plt.title('Line Plot 4')
plt.tight_layout()

# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a graphical representation of data in which bars of rectangular shapes are used to depict the values of different categories. The length (or height) of each bar corresponds to the quantity or frequency of the data it represents. Bar plots are commonly used to display and compare the values of different categories or groups.

Key features of a bar plot:

Categories on one axis, values on the other: The categories (or groups) are typically represented on one axis (usually the x-axis), and the corresponding values are represented on the other axis (usually the y-axis).

Discrete data: Bar plots are suitable for visualizing discrete data, where each category is separate and distinct.

Why is it used:

Comparison of Categories: Bar plots are effective for comparing the quantities or frequencies of different categories. For example, you might use a bar plot to compare sales figures for different products, populations of different cities, or scores of different students.

Categorical Data Visualization: They are particularly useful for visualizing categorical data, where the data points can be grouped into distinct categories or groups.

Readability: Bar plots are easy to read and interpret, making them a popular choice for conveying information to a broad audience.

Simple Representation: They provide a straightforward representation of data without the complexities often associated with other types of plots.

Variations: Bar plots can take different forms, such as grouped bar plots, stacked bar plots, and horizontal bar plots, providing flexibility in visualizing various types of data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2, 2, 1)  
plt.bar(company,profit,color='r')
plt.subplot(2, 2, 2) 
plt.barh(company,profit,color='r')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)


Box Plot:

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It is particularly useful for displaying the spread and skewness of the data, as well as identifying outliers. The box plot is based on percentiles and quartiles, and it consists of several key components:

Box: The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The length of the box indicates the spread of the middle 50% of the data.

Line (whisker): The lines (whiskers) extending from the box indicate the range of the data, excluding outliers. These lines can extend to a certain multiple of the IQR, and points beyond the whiskers are often considered outliers.

Median: A line inside the box represents the median (Q2), which is the middle value of the dataset when it is ordered.

Outliers: Individual data points that fall outside the whiskers are considered outliers and are often plotted individually.

Why is it used:

Data Distribution: Box plots provide a visual summary of the distribution of a dataset, highlighting key statistics such as the median, quartiles, and potential outliers.

Comparison of Distributions: They are useful for comparing the distributions of different datasets or groups. Multiple box plots can be displayed side by side for easy comparison.

Identifying Outliers: Outliers, which may indicate errors or interesting phenomena in the data, are visually identified in a box plot.

Skewness: Box plots provide insights into the skewness of the data. A skewed distribution may have a longer tail on one side.

Compact Representation: Despite their simplicity, box plots efficiently convey a lot of information about the central tendency and spread of the data in a compact form.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
plt.subplot(2, 2, 1) 
plt.boxplot(box1)
plt.subplot(2, 2, 2) 
plt.boxplot(box2)
```
