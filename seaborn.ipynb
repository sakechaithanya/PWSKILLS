Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a data visualization library for Python that is based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five types of plots that you can create using Seaborn, along with their typical uses:

Scatter Plot:

Use: Scatter plots are useful for visualizing the relationship between two continuous variables. Each point on the plot represents an observation in your dataset.
Seaborn Function: seaborn.scatterplot()
Line Plot:

Use: Line plots are effective for displaying trends over a continuous interval or time. They are suitable for visualizing changes in a variable over a range.
Seaborn Function: seaborn.lineplot()
Histogram:

Use: Histograms are used to visualize the distribution of a single continuous variable. They display the frequencies of different ranges or bins.
Seaborn Function: seaborn.histplot()
Box Plot (Box-and-Whisker Plot):

Use: Box plots are helpful in visualizing the distribution of a continuous variable across different categories. They show the median, quartiles, and potential outliers.
Seaborn Function: seaborn.boxplot()
Heatmap:

Use: Heatmaps are used for displaying the relationships between two categorical variables. They are particularly effective in showing the concentration of data points in a matrix format.
Seaborn Function: seaborn.heatmap()

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
frmi_data=sns.load_dataset( "fmri")
```

```python
frmi_data
```

```python
sns.lineplot(x='timepoint',y='signal',data= fmri_data,hue='event',style='region')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data=sns.load_dataset('flights')
```

```python
flights_data
```

```python
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(flights_pivot, annot=True, cmap='viridis', fmt='d', linewidths=.5)
```
